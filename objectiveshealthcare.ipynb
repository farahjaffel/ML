#fine-tune healthcare management



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report, confusion_matrix, mean_squared_error, r2_score
from sklearn.cluster import KMeans
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import os
import joblib
import warnings
warnings.filterwarnings('ignore')

class HealthcareManagementModels:
    """
    A class for developing models to fine-tune healthcare management.
    Implements 4 model types:
    1. Supervised Classification
    2. Supervised Regression
    3. Unsupervised Clustering
    4. Time Series (LSTM)
    """

    def __init__(self):
        """Initialize the class with empty attributes."""
        self.healthcare_data = None
        self.hospital_data = None
        self.merged_data = None
        self.classification_model = None
        self.regression_model = None
        self.clustering_model = None
        self.time_series_model = None

    def load_data(self, healthcare_path=None, hospital_path=None, use_sample_data=False):
        """
        Load and preprocess all required datasets.

        Parameters:
        - healthcare_path: path to healthcare data CSV
        - hospital_path: path to hospital data CSV
        - use_sample_data: if True, generate sample data instead of loading from files
        """
        print("Loading and preprocessing data...")

        if use_sample_data:
            print("Using generated sample data instead of loading from files")
            self._generate_sample_data()
            return True

        # Load datasets
        try:
            healthcare_path = healthcare_path or '/content/merged_healthcare_data.csv'
            hospital_path = hospital_path or '/content/merged_hospital_admission_data.csv'

            if os.path.exists(healthcare_path):
                self.healthcare_data = pd.read_csv(healthcare_path)
                print(f"Healthcare data loaded: {self.healthcare_data.shape}")
            else:
                print(f"Healthcare data file not found at: {healthcare_path}")
                print("Generating sample healthcare data instead")
                self._generate_sample_healthcare_data()

            if os.path.exists(hospital_path):
                self.hospital_data = pd.read_csv(hospital_path)
                print(f"Hospital data loaded: {self.hospital_data.shape}")
            else:
                print(f"Hospital data file not found at: {hospital_path}")
                print("Generating sample hospital data instead")
                self._generate_sample_hospital_data()

            # Merge datasets on subject_id if available
            if self.healthcare_data is not None and self.hospital_data is not None:
                if 'subject_id' in self.healthcare_data.columns and 'subject_id' in self.hospital_data.columns:
                    self.merged_data = pd.merge(
                        self.healthcare_data,
                        self.hospital_data,
                        on='subject_id',
                        how='inner',
                        suffixes=('', '_hospital')
                    )
                    print(f"Merged data: {self.merged_data.shape}")
                else:
                    print("Cannot merge datasets - subject_id not found in both datasets")
                    self.merged_data = self.healthcare_data.copy()
            else:
                # If either dataset is missing, use the one that exists
                self.merged_data = self.healthcare_data if self.healthcare_data is not None else self.hospital_data

            # Handle missing values
            self._handle_missing_values()

            # Feature engineering
            self._engineer_features()

            return True

        except Exception as e:
            print(f"Error loading data: {e}")
            print("Falling back to sample data")
            self._generate_sample_data()
            return True

    def _generate_sample_data(self):
        """Generate sample data for both healthcare and hospital datasets."""
        self._generate_sample_healthcare_data()
        self._generate_sample_hospital_data()

        # Merge the data
        if 'subject_id' in self.healthcare_data.columns and 'subject_id' in self.hospital_data.columns:
            self.merged_data = pd.merge(
                self.healthcare_data,
                self.hospital_data,
                on='subject_id',
                how='inner',
                suffixes=('', '_hospital')
            )
            print(f"Merged data: {self.merged_data.shape}")
        else:
            print("Cannot merge datasets - subject_id not found in both datasets")
            self.merged_data = self.healthcare_data.copy()

        # Handle missing values
        self._handle_missing_values()

        # Feature engineering
        self._engineer_features()

    def _generate_sample_healthcare_data(self):
        """Generate sample healthcare data."""
        # Set random seed for reproducibility
        np.random.seed(42)

        # Number of samples
        n_samples = 200

        # Generate subject IDs
        subject_ids = np.arange(1, n_samples + 1)

        # Generate complaint types
        complaint_types = np.random.choice(
            ['Billing', 'Clinical', 'Administrative', 'Facility', 'Staff'],
            size=n_samples
        )

        # Generate resolution durations (1-30 days)
        resolution_durations = np.random.gamma(shape=2.0, scale=3.0, size=n_samples)
        resolution_durations = np.round(np.clip(resolution_durations, 1, 30))

        # Generate satisfaction impacts (1-10 scale)
        satisfaction_impacts = np.random.beta(a=2, b=1.5, size=n_samples) * 9 + 1
        satisfaction_impacts = np.round(satisfaction_impacts, 1)

        # Generate dates
        base_date = pd.Timestamp('2023-01-01')
        dates = [base_date + pd.Timedelta(days=int(np.random.uniform(0, 365))) for _ in range(n_samples)]

        # Generate gender
        genders = np.random.choice(['M', 'F', 'Other'], size=n_samples, p=[0.48, 0.48, 0.04])

        # Generate marital status
        marital_statuses = np.random.choice(
            ['Single', 'Married', 'Divorced', 'Widowed'],
            size=n_samples,
            p=[0.3, 0.5, 0.15, 0.05]
        )

        # Generate BMI
        bmis = np.random.normal(loc=25, scale=5, size=n_samples)
        bmis = np.round(np.clip(bmis, 15, 45), 1)

        # Create DataFrame
        self.healthcare_data = pd.DataFrame({
            'subject_id': subject_ids,
            'Complaint_Type': complaint_types,
            'Resolution_Duration_Days': resolution_durations,
            'Satisfaction_Impact': satisfaction_impacts,
            'Complaint_Date': dates,
            'gender': genders,
            'MARITAL_STATUS': marital_statuses,
            'BMI': bmis
        })

        print(f"Sample healthcare data generated: {self.healthcare_data.shape}")

    def _generate_sample_hospital_data(self):
        """Generate sample hospital data."""
        # Set random seed for reproducibility
        np.random.seed(43)

        # Number of samples (more than healthcare data to simulate only some having complaints)
        n_samples = 300

        # Generate subject IDs (ensure overlap with healthcare data)
        if self.healthcare_data is not None:
            # Ensure all healthcare subjects are included
            healthcare_ids = set(self.healthcare_data['subject_id'].values)
            n_new = n_samples - len(healthcare_ids)
            new_ids = np.arange(max(healthcare_ids) + 1, max(healthcare_ids) + n_new + 1)
            subject_ids = np.concatenate([np.array(list(healthcare_ids)), new_ids])
            np.random.shuffle(subject_ids)
            subject_ids = subject_ids[:n_samples]
        else:
            subject_ids = np.arange(1, n_samples + 1)

        # Generate length of stay (days)
        length_of_stay = np.random.gamma(shape=1.5, scale=4.0, size=n_samples)
        length_of_stay = np.round(np.clip(length_of_stay, 1, 60))

        # Generate total cost
        base_cost = 5000
        cost_per_day = np.random.uniform(800, 2000, size=n_samples)
        total_cost = base_cost + (length_of_stay * cost_per_day)

        # Generate admission dates
        base_date = pd.Timestamp('2023-01-01')
        admission_dates = [base_date + pd.Timedelta(days=int(np.random.uniform(0, 300))) for _ in range(n_samples)]

        # Create DataFrame
        self.hospital_data = pd.DataFrame({
            'subject_id': subject_ids,
            'length_of_stay': length_of_stay,
            'total_cost': total_cost,
            'admission_date': admission_dates
        })

        print(f"Sample hospital data generated: {self.hospital_data.shape}")

    def _handle_missing_values(self):
        """Handle missing values in the data."""
        # For numerical columns, fill with median
        for df in [self.healthcare_data, self.hospital_data, self.merged_data]:
            if df is not None:
                num_cols = df.select_dtypes(include=['int64', 'float64']).columns
                for col in num_cols:
                    if df[col].isnull().sum() > 0:
                        df[col] = df[col].fillna(df[col].median())

                # For categorical columns, fill with mode
                cat_cols = df.select_dtypes(include=['object']).columns
                for col in cat_cols:
                    if df[col].isnull().sum() > 0:
                        df[col] = df[col].fillna(df[col].mode()[0] if not df[col].mode().empty else 'Unknown')

    def _engineer_features(self):
        """Create new features for the models."""
        # Process healthcare data if available
        if self.healthcare_data is not None:
            # Categorize satisfaction if present
            if 'Satisfaction_Impact' in self.healthcare_data.columns:
                self.healthcare_data['satisfaction_category'] = pd.cut(
                    self.healthcare_data['Satisfaction_Impact'],
                    bins=[0, 3, 7, 10],
                    labels=['Low', 'Medium', 'High']
                )

            # Categorize resolution time if present
            if 'Resolution_Duration_Days' in self.healthcare_data.columns:
                self.healthcare_data['resolution_category'] = pd.cut(
                    self.healthcare_data['Resolution_Duration_Days'],
                    bins=[0, 3, 7, float('inf')],
                    labels=['Fast', 'Medium', 'Slow']
                )

            # Process dates if available
            date_cols = [col for col in self.healthcare_data.columns if 'date' in col.lower() or 'Date' in col]
            for col in date_cols:
                try:
                    self.healthcare_data[col] = pd.to_datetime(self.healthcare_data[col])
                except:
                    pass

        # Process hospital data if available
        if self.hospital_data is not None:
            # Convert length_of_stay to numeric if present
            if 'length_of_stay' in self.hospital_data.columns:
                try:
                    self.hospital_data['length_of_stay'] = pd.to_numeric(self.hospital_data['length_of_stay'], errors='coerce')
                except:
                    pass

            # Categorize length of stay if present
            if 'length_of_stay' in self.hospital_data.columns:
                self.hospital_data['stay_category'] = pd.cut(
                    pd.to_numeric(self.hospital_data['length_of_stay'], errors='coerce'),
                    bins=[0, 3, 7, float('inf')],
                    labels=['Short', 'Medium', 'Long']
                )

        # Process merged data if available
        if self.merged_data is not None:
            # Make sure all needed columns are present
            needed_columns = ['Satisfaction_Impact', 'Resolution_Duration_Days']
            if not all(col in self.merged_data.columns for col in needed_columns):
                missing_cols = [col for col in needed_columns if col not in self.merged_data.columns]
                print(f"Warning: Missing columns in merged data: {missing_cols}")

            # Add satisfaction category if not present
            if 'satisfaction_category' not in self.merged_data.columns and 'Satisfaction_Impact' in self.merged_data.columns:
                self.merged_data['satisfaction_category'] = pd.cut(
                    self.merged_data['Satisfaction_Impact'],
                    bins=[0, 3, 7, 10],
                    labels=['Low', 'Medium', 'High']
                )

            # Add resolution category if not present
            if 'resolution_category' not in self.merged_data.columns and 'Resolution_Duration_Days' in self.merged_data.columns:
                self.merged_data['resolution_category'] = pd.cut(
                    self.merged_data['Resolution_Duration_Days'],
                    bins=[0, 3, 7, float('inf')],
                    labels=['Fast', 'Medium', 'Slow']
                )

            # Add stay category if not present
            if 'stay_category' not in self.merged_data.columns and 'length_of_stay' in self.merged_data.columns:
                self.merged_data['stay_category'] = pd.cut(
                    pd.to_numeric(self.merged_data['length_of_stay'], errors='coerce'),
                    bins=[0, 3, 7, float('inf')],
                    labels=['Short', 'Medium', 'Long']
                )

            # Create combined features if possible
            if 'Satisfaction_Impact' in self.merged_data.columns and 'total_cost' in self.merged_data.columns:
                try:
                    self.merged_data['cost_per_satisfaction'] = pd.to_numeric(self.merged_data['total_cost'], errors='coerce') / self.merged_data['Satisfaction_Impact']
                except:
                    pass

    def build_classification_model(self):
        """
        Build a classification model to predict satisfaction categories.
        Returns the trained model, preprocessor, and evaluation metrics.
        """
        print("\n=== Building Classification Model for Healthcare Management ===")

        # Use healthcare data or merged data if available
        data = self.merged_data if self.merged_data is not None else self.healthcare_data

        # Ensure data is loaded
        if data is None:
            print("Error: No data available. Load data first.")
            return None

    def save_models(self, save_dir="saved_models"):
        """
        Save all trained models to the specified directory.

        Parameters:
        - save_dir: str, directory where models will be saved (default: 'saved_models')

        Returns:
        - dict: mapping model names to their save paths (or None if not saved)
        """
        print(f"Saving models to {save_dir}...")

        # Create save directory if it doesn't exist
        os.makedirs(save_dir, exist_ok=True)

        saved_paths = {
            'classification': None,
            'regression': None,
            'clustering': None,
            'time_series': None
        }

        # Save classification model
        if self.classification_model is not None:
            try:
                class_path = os.path.join(save_dir, "classification_model.joblib")
                joblib.dump(self.classification_model['model'], class_path)
                print(f"Classification model saved to {class_path}")
                saved_paths['classification'] = class_path
            except Exception as e:
                print(f"Error saving classification model: {e}")

        # Save regression model
        if self.regression_model is not None:
            try:
                reg_path = os.path.join(save_dir, "regression_model.joblib")
                joblib.dump(self.regression_model['model'], reg_path)
                print(f"Regression model saved to {reg_path}")
                saved_paths['regression'] = reg_path
            except Exception as e:
                print(f"Error saving regression model: {e}")

        # Save clustering model (KMeans and scaler)
        if self.clustering_model is not None:
            try:
                # Save KMeans model
                cluster_path = os.path.join(save_dir, "clustering_model.joblib")
                joblib.dump(self.clustering_model['model'], cluster_path)
                print(f"Clustering model saved to {cluster_path}")
                saved_paths['clustering'] = cluster_path

                # Save scaler (needed for clustering predictions)
                scaler_path = os.path.join(save_dir, "clustering_scaler.joblib")
                joblib.dump(self.clustering_model['scaler'], scaler_path)
                print(f"Clustering scaler saved to {scaler_path}")
            except Exception as e:
                print(f"Error saving clustering model: {e}")

        # Save time series model
        if self.time_series_model is not None:
            print("Attempting to save time series model...")
            try:
                # First try saving as .keras
                ts_path = os.path.join(save_dir, "time_series_model.keras")
                print(f"Saving time series model to {ts_path}...")
                self.time_series_model['model'].save(ts_path)
                print(f"Time series model saved to {ts_path}")
                saved_paths['time_series'] = ts_path

                # Save scaler (needed for time series predictions)
                ts_scaler_path = os.path.join(save_dir, "time_series_scaler.joblib")
                print(f"Saving time series scaler to {ts_scaler_path}...")
                joblib.dump(self.time_series_model['scaler'], ts_scaler_path)
                print(f"Time series scaler saved to {ts_scaler_path}")
            except Exception as e:
                print(f"Error saving time series model with .keras: {e}")
                # Fallback to .h5 format
                try:
                    ts_path = os.path.join(save_dir, "time_series_model.h5")
                    print(f"Fallback: Saving time series model to {ts_path}...")
                    self.time_series_model['model'].save(ts_path, save_format='h5')
                    print(f"Time series model saved to {ts_path}")
                    saved_paths['time_series'] = ts_path

                    # Save scaler
                    ts_scaler_path = os.path.join(save_dir, "time_series_scaler.joblib")
                    print(f"Saving time series scaler to {ts_scaler_path}...")
                    joblib.dump(self.time_series_model['scaler'], ts_scaler_path)
                    print(f"Time series scaler saved to {ts_scaler_path}")
                except Exception as e2:
                    print(f"Error saving time series model with .h5: {e2}")
                    saved_paths['time_series'] = None
        else:
            print("Time series model is None, skipping save.")

        return saved_paths

    def load_models(self, save_dir="saved_models"):
        """
        Load all saved models from the specified directory.

        Parameters:
        - save_dir: str, directory where models are saved (default: 'saved_models')

        Returns:
        - dict: mapping model names to loaded models (or None if not loaded)
        """
        print(f"Loading models from {save_dir}...")

        loaded_models = {
            'classification': None,
            'regression': None,
            'clustering': None,
            'time_series': None
        }

        # Load classification model
        class_path = os.path.join(save_dir, "classification_model.joblib")
        if os.path.exists(class_path):
            try:
                self.classification_model = {
                    'model': joblib.load(class_path),
                    'features': self.classification_model['features'] if self.classification_model else [],
                    'accuracy': self.classification_model['accuracy'] if self.classification_model else None,
                    'target': self.classification_model['target'] if self.classification_model else None,
                    'classes': self.classification_model['classes'] if self.classification_model else None
                }
                print(f"Classification model loaded from {class_path}")
                loaded_models['classification'] = self.classification_model
            except Exception as e:
                print(f"Error loading classification model: {e}")

        # Load regression model
        reg_path = os.path.join(save_dir, "regression_model.joblib")
        if os.path.exists(reg_path):
            try:
                self.regression_model = {
                    'model': joblib.load(reg_path),
                    'features': self.regression_model['features'] if self.regression_model else [],
                    'mse': self.regression_model['mse'] if self.regression_model else None,
                    'r2': self.regression_model['r2'] if self.regression_model else None,
                    'target': self.regression_model['target'] if self.regression_model else None
                }
                print(f"Regression model loaded from {reg_path}")
                loaded_models['regression'] = self.regression_model
            except Exception as e:
                print(f"Error loading regression model: {e}")

        # Load clustering model
        cluster_path = os.path.join(save_dir, "clustering_model.joblib")
        scaler_path = os.path.join(save_dir, "clustering_scaler.joblib")
        if os.path.exists(cluster_path) and os.path.exists(scaler_path):
            try:
                self.clustering_model = {
                    'model': joblib.load(cluster_path),
                    'scaler': joblib.load(scaler_path),
                    'features': self.clustering_model['features'] if self.clustering_model else [],
                    'cluster_analysis': self.clustering_model['cluster_analysis'] if self.clustering_model else None
                }
                print(f"Clustering model loaded from {cluster_path}")
                print(f"Clustering scaler loaded from {scaler_path}")
                loaded_models['clustering'] = self.clustering_model
            except Exception as e:
                print(f"Error loading clustering model: {e}")

        # Load time series model
        ts_path_keras = os.path.join(save_dir, "time_series_model.keras")
        ts_path_h5 = os.path.join(save_dir, "time_series_model.h5")
        ts_scaler_path = os.path.join(save_dir, "time_series_scaler.joblib")

        # Try loading .keras first, then .h5
        ts_path = None
        if os.path.exists(ts_path_keras):
            ts_path = ts_path_keras
        elif os.path.exists(ts_path_h5):
            ts_path = ts_path_h5

        if ts_path and os.path.exists(ts_scaler_path):
            try:
                self.time_series_model = {
                    'model': tf.keras.models.load_model(ts_path),
                    'scaler': joblib.load(ts_scaler_path),
                    'seq_length': self.time_series_model['seq_length'] if self.time_series_model else None,
                    'forecast': self.time_series_model['forecast'] if self.time_series_model else None,
                    'forecast_dates': self.time_series_model['forecast_dates'] if self.time_series_model else None
                }
                print(f"Time series model loaded from {ts_path}")
                print(f"Time series scaler loaded from {ts_scaler_path}")
                loaded_models['time_series'] = self.time_series_model
            except Exception as e:
                print(f"Error loading time series model: {e}")

        return loaded_models

        # Check if target is available
        if 'satisfaction_category' not in data.columns:
            print("Error: Target variable 'satisfaction_category' not found")
            return None

        # Define features and target
        target = 'satisfaction_category'

        # Select categorical features
        cat_features = ['Complaint_Type', 'MARITAL_STATUS', 'gender']
        cat_features = [f for f in cat_features if f in data.columns]

        # Select numerical features
        num_features = ['Resolution_Duration_Days', 'Satisfaction_Impact']
        additional_features = ['total_cost', 'length_of_stay', 'BMI']
        num_features.extend([f for f in additional_features if f in data.columns])
        num_features = [f for f in num_features if f != target]  # Remove target if in features

        print(f"Using categorical features: {cat_features}")
        print(f"Using numerical features: {num_features}")

        # Check if we have enough features
        if not cat_features and not num_features:
            print("Error: No valid features found")
            return None

        # Create preprocessing pipeline
        preprocessor = ColumnTransformer(
            transformers=[
                ('num', StandardScaler(), num_features),
                ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)
            ]
        )

        # Create model pipeline
        pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
        ])

        # Filter out rows with missing target
        data_model = data.dropna(subset=[target])

        # Split data
        X = data_model[cat_features + num_features]
        y = data_model[target]

        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

        print(f"Training data shape: {X_train.shape}")
        print(f"Testing data shape: {X_test.shape}")

        # Train model
        pipeline.fit(X_train, y_train)

        # Evaluate model
        y_pred = pipeline.predict(X_test)
        accuracy = (y_pred == y_test).mean()

        print(f"Model Accuracy: {accuracy:.4f}")
        print("\nClassification Report:")
        print(classification_report(y_test, y_pred))

        # Plot confusion matrix
        plt.figure(figsize=(8, 6))
        cm = confusion_matrix(y_test, y_pred)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=pipeline.classes_, yticklabels=pipeline.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix - Satisfaction Classification')
        plt.tight_layout()
        plt.show()

        # Save the model
        self.classification_model = {
            'model': pipeline,
            'features': cat_features + num_features,
            'accuracy': accuracy,
            'target': target,
            'classes': pipeline.classes_
        }

        return self.classification_model

    def build_regression_model(self):
        """
        Build a regression model to predict exact length of stay.
        Returns the trained model, preprocessor, and evaluation metrics.
        """
        print("\n=== Building Regression Model for Treatment Duration ===")

        # Use hospital data or merged data if available
        data = self.merged_data if self.merged_data is not None else self.hospital_data

        # Check if target is available
        if 'length_of_stay' not in data.columns:
            print("Error: Target variable 'length_of_stay' not found")
            return None

        # Define features and target
        target = 'length_of_stay'

        # Select categorical features (same as classification for consistency)
        cat_features = ['DIAGNOSIS', 'diagnosis_category', 'ADMISSION_TYPE', 'gender', 'ETHNICITY']
        cat_features = [f for f in cat_features if f in data.columns]

        # Select numerical features, checking for existence
        possible_num_features = ['BMI', 'comorbidity_count', 'total_cost', 'HadDiabetes', 'age']
        num_features = [f for f in possible_num_features if f in data.columns]

        print(f"Using categorical features: {cat_features}")
        print(f"Using numerical features: {num_features}")

        # Check if we have enough features
        if not cat_features and not num_features:
            print("Error: No valid features found")
            return None

        # Filter out rows with missing target
        data_model = data.dropna(subset=[target])

        # Check if data_model has enough rows
        if len(data_model) == 0:
            print("Error: No valid data available after removing missing 'length_of_stay' values")
            return None

        # Split data
        X = data_model[cat_features + num_features]
        y = data_model[target]

        # Check for sufficient samples
        if len(X) < 2:
            print("Error: Insufficient data samples to perform train-test split")
            return None

        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

        # Create preprocessing pipeline
        preprocessor = ColumnTransformer(
            transformers=[
                ('num', StandardScaler(), num_features),
                ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)
            ]
        )

        # Create model pipeline with Gradient Boosting Regressor
        pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('regressor', GradientBoostingRegressor(random_state=42))
        ])

        # Train model
        pipeline.fit(X_train, y_train)

        # Evaluate model
        y_pred = pipeline.predict(X_test)
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

        print(f"Mean Squared Error: {mse:.4f}")
        print(f"Root Mean Squared Error: {np.sqrt(mse):.4f}")
        print(f"R-squared: {r2:.4f}")

        # Feature importance
        if hasattr(pipeline['regressor'], 'feature_importances_'):
            # Get feature names after one-hot encoding
            ohe = pipeline.named_steps['preprocessor'].transformers_[1][1]
            cat_feature_names = []
            if cat_features:
                cat_feature_names = ohe.get_feature_names_out(cat_features).tolist()

            feature_names = num_features + cat_feature_names

            # Map importances to feature names
            importances = pipeline['regressor'].feature_importances_
            if len(importances) == len(feature_names):
                feature_importance = pd.DataFrame({
                    'Feature': feature_names,
                    'Importance': importances
                }).sort_values('Importance', ascending=False)

                print("\nTop 10 Important Features:")
                print(feature_importance.head(10))

                # Plot feature importance
                plt.figure(figsize=(12, 8))
                sns.barplot(x='Importance', y='Feature', data=feature_importance.head(10))
                plt.title('Feature Importance for Length of Stay Prediction')
                plt.tight_layout()
                plt.show()

        # Plot actual vs predicted
        plt.figure(figsize=(10, 6))
        plt.scatter(y_test, y_pred, alpha=0.5)
        plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
        plt.xlabel('Actual Length of Stay')
        plt.ylabel('Predicted Length of Stay')
        plt.title('Actual vs Predicted Length of Stay')
        plt.tight_layout()
        plt.show()

        # Save the model
        self.regression_model = {
            'model': pipeline,
            'features': cat_features + num_features,
            'mse': mse,
            'r2': r2,
            'target': target
        }

        return self.regression_model

    def build_clustering_model(self):
        """
        Build a clustering model to identify patient satisfaction patterns.
        Returns the trained model and cluster analysis.
        """
        print("\n=== Building Clustering Model for Healthcare Management ===")

        # Use healthcare data or merged data if available
        data = self.merged_data if self.merged_data is not None else self.healthcare_data

        # Ensure data is loaded
        if data is None:
            print("Error: No data available. Load data first.")
            return None

        # Select features for clustering
        cluster_features = ['Satisfaction_Impact', 'Resolution_Duration_Days']
        additional_features = ['total_cost', 'length_of_stay']

        # Add available additional features
        cluster_features.extend([f for f in additional_features if f in data.columns])

        # Check if we have the necessary features
        available_features = [f for f in cluster_features if f in data.columns]

        if len(available_features) < 2:
            print("Error: Not enough features available for clustering")
            return None

        print(f"Using features for clustering: {available_features}")

        # Prepare data for clustering
        X = data[available_features].copy()

        # Convert to numeric if needed
        for col in X.columns:
            if X[col].dtype == 'object':
                X[col] = pd.to_numeric(X[col], errors='coerce')

        # Drop rows with missing values
        X = X.dropna()

        if len(X) < 10:
            print("Error: Not enough complete data points for clustering")
            return None

        # Scale the features
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)

        # Determine optimal number of clusters using the elbow method
        inertia = []
        k_range = range(1, min(6, len(X)))

        for k in k_range:
            kmeans = KMeans(n_clusters=k, random_state=42)
            kmeans.fit(X_scaled)
            inertia.append(kmeans.inertia_)

        # Plot elbow curve
        plt.figure(figsize=(10, 6))
        plt.plot(k_range, inertia, 'o-')
        plt.xlabel('Number of Clusters')
        plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
        plt.title('Elbow Method for Optimal k')
        plt.tight_layout()
        plt.show()

        # Choose optimal number of clusters (simplified method)
        k_optimal = 3  # Default to 3 clusters, could be determined from elbow curve

        # Build final clustering model
        kmeans = KMeans(n_clusters=k_optimal, random_state=42)
        cluster_labels = kmeans.fit_predict(X_scaled)

        # Add cluster labels to original data
        X['cluster'] = cluster_labels

        # Analyze clusters
        cluster_analysis = X.groupby('cluster').mean()
        print("\nCluster Analysis:")
        print(cluster_analysis)

        # Visualize clusters (first two features)
        plt.figure(figsize=(12, 8))

        if len(available_features) >= 2:
            scatter = plt.scatter(
                X_scaled[:, 0],
                X_scaled[:, 1],
                c=cluster_labels,
                cmap='viridis',
                alpha=0.8
            )

            plt.colorbar(scatter, label='Cluster')
            plt.xlabel(available_features[0])
            plt.ylabel(available_features[1])
            plt.title('Patient Satisfaction Clusters')

            # Add cluster centers
            centers = kmeans.cluster_centers_
            plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Cluster Centers')
            plt.legend()

            plt.tight_layout()
            plt.show()

        # Save the model
        self.clustering_model = {
            'model': kmeans,
            'scaler': scaler,
            'features': available_features,
            'cluster_analysis': cluster_analysis
        }

        return self.clustering_model

    def run_all_models(self):
        """
        Run all models (classification, regression, clustering, time series) with the loaded data.
        Returns a dictionary with results from each model.
        """
        print("Running all models...")

        # Ensure data is loaded
        if self.merged_data is None and self.healthcare_data is None:
            print("No data available, loading sample data...")
            self.load_data(use_sample_data=True)

        results = {}

        # Run classification model
        print("\nRunning Classification Model...")
        classification_result = self.build_classification_model()
        results['classification'] = classification_result

        # Run regression model
        print("\nRunning Regression Model...")
        regression_result = self.build_regression_model()
        results['regression'] = regression_result

        # Run clustering model
        print("\nRunning Clustering Model...")
        clustering_result = self.build_clustering_model()
        results['clustering'] = clustering_result

        # Run time series model
        print("\nRunning Time Series Model...")
        time_series_result = self.build_time_series_model()
        results['time_series'] = time_series_result

        return results

    def generate_insights(self):
        """
        Generate insights based on the results of all models.
        Returns a dictionary with key insights.
        """
        print("Generating insights from models...")

        insights = {}

        # Classification insights
        if self.classification_model:
            insights['classification'] = {
                'accuracy': self.classification_model['accuracy'],
                'features_used': self.classification_model['features'],
                'note': "Model predicts satisfaction categories (Low, Medium, High) based on complaint types and patient characteristics."
            }

        # Regression insights
        if self.regression_model:
            insights['regression'] = {
                'r2_score': self.regression_model['r2'],
                'mse': self.regression_model['mse'],
                'features_used': self.regression_model['features'],
                'note': "Model predicts continuous satisfaction scores, useful for understanding impact of resolution time and costs."
            }

        # Clustering insights
        if self.clustering_model:
            cluster_means = self.clustering_model['cluster_analysis'].to_dict()
            insights['clustering'] = {
                'features_used': self.clustering_model['features'],
                'cluster_means': cluster_means,
                'note': "Identifies patient groups with similar satisfaction patterns, useful for targeted interventions."
            }

        # Time series insights
        if self.time_series_model:
            forecast_mean = float(np.mean(self.time_series_model['forecast'])) if self.time_series_model['forecast'] is not None else None
            insights['time_series'] = {
                'forecast_mean': forecast_mean,
                'sequence_length': self.time_series_model['seq_length'],
                'note': "Predicts future satisfaction trends, useful for planning resource allocation."
            }

        return insights

    def build_time_series_model(self):
        """
        Build a time series model to predict satisfaction trends.
        Returns the trained LSTM model and forecast.
        """
        print("\n=== Building Time Series Model for Healthcare Management ===")

        # Use healthcare data or merged data if available
        data = self.merged_data if self.merged_data is not None else self.healthcare_data

        # Ensure data is loaded
        if data is None:
            print("Error: No data available. Load data first.")
            return None

        # Check if we have date and satisfaction information
        date_cols = [col for col in data.columns if 'date' in col.lower() or 'Date' in col]

        if not date_cols or 'Satisfaction_Impact' not in data.columns:
            print("Error: Missing date column or 'Satisfaction_Impact' column")
            return None

        # Select date column to use
        date_col = date_cols[0]
        print(f"Using date column: {date_col}")

        # Create time series data
        try:
            # Convert to datetime
            data[date_col] = pd.to_datetime(data[date_col])

            # Sort by date and create time series
            ts_data = data.sort_values(date_col)

            # Create daily time series (resample if enough data points)
            if len(ts_data) > 30:
                # Set date as index
                ts_data = ts_data.set_index(date_col)

                # Resample to daily and fill missing values
                daily_satisfaction = ts_data['Satisfaction_Impact'].resample('D').mean().fillna(method='ffill')

                # For demonstration, if we don't have enough data points, we'll generate synthetic data
                if len(daily_satisfaction) < 60:
                    # Get date range
                    start_date = daily_satisfaction.index.min()
                    end_date = daily_satisfaction.index.max() + pd.Timedelta(days=60)

                    # Create synthetic time series with similar characteristics
                    dates = pd.date_range(start=start_date, end=end_date, freq='D')
                    values = np.random.normal(
                        loc=daily_satisfaction.mean(),
                        scale=daily_satisfaction.std(),
                        size=len(dates)
                    )

                    # Add trend and seasonality
                    trend = np.linspace(0, 1, len(dates)) * 0.5  # Upward trend
                    seasonality = 0.2 * np.sin(np.linspace(0, 10*np.pi, len(dates)))  # Seasonal pattern

                    values = values + trend + seasonality

                    # Ensure values are in reasonable range (1-10)
                    values = np.clip(values, 1, 10)

                    # Create time series
                    daily_satisfaction = pd.Series(values, index=dates)

                    print("Note: Using synthetic data to demonstrate time series modeling")
            else:
                print("Error: Not enough data points for time series analysis")
                return None

            # Plot the time series
            plt.figure(figsize=(12, 6))
            daily_satisfaction.plot()
            plt.title('Satisfaction Score Time Series')
            plt.xlabel('Date')
            plt.ylabel('Satisfaction Score')
            plt.tight_layout()
            plt.show()

            # Prepare data for LSTM
            def create_sequences(data, seq_length):
                """Create sequences for LSTM model"""
                xs, ys = [], []
                for i in range(len(data) - seq_length):
                    x = data[i:i+seq_length]
                    y = data[i+seq_length]
                    xs.append(x)
                    ys.append(y)
                return np.array(xs), np.array(ys)

            # Normalize the data
            scaler = StandardScaler()
            satisfaction_scaled = scaler.fit_transform(daily_satisfaction.values.reshape(-1, 1))

            # Define sequence length (lookback period)
            seq_length = 7  # One week lookback

            # Create sequences
            X, y = create_sequences(satisfaction_scaled, seq_length)

            # Reshape for LSTM [samples, time steps, features]
            X = X.reshape(X.shape[0], X.shape[1], 1)

            # Split into train and test
            split_ratio = 0.8
            split_index = int(len(X) * split_ratio)

            X_train, X_test = X[:split_index], X[split_index:]
            y_train, y_test = y[:split_index], y[split_index:]

            print(f"Training sequences: {X_train.shape}")
            print(f"Testing sequences: {X_test.shape}")

            # Build LSTM model
            model = Sequential([
                LSTM(50, activation='relu', input_shape=(seq_length, 1), return_sequences=True),
                Dropout(0.2),
                LSTM(50, activation='relu'),
                Dropout(0.2),
                Dense(1)
            ])

            model.compile(optimizer='adam', loss='mse')

            # Train model with early stopping
            early_stop = tf.keras.callbacks.EarlyStopping(
                monitor='val_loss',
                patience=10,
                restore_best_weights=True
            )

            history = model.fit(
                X_train, y_train,
                epochs=100,
                batch_size=16,
                validation_split=0.1,
                callbacks=[early_stop],
                verbose=1
            )

            # Plot training history
            plt.figure(figsize=(10, 6))
            plt.plot(history.history['loss'], label='Train Loss')
            plt.plot(history.history['val_loss'], label='Validation Loss')
            plt.title('Model Loss')
            plt.ylabel('Loss')
            plt.xlabel('Epoch')
            plt.legend()
            plt.tight_layout()
            plt.show()

            # Evaluate model
            y_pred = model.predict(X_test)
            mse = mean_squared_error(y_test, y_pred)
            print(f"Test MSE: {mse:.4f}")

            # Inverse transform predictions
            y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))
            y_pred_actual = scaler.inverse_transform(y_pred)

            # Plot test predictions
            plt.figure(figsize=(12, 6))
            plt.plot(y_test_actual, label='Actual')
            plt.plot(y_pred_actual, label='Predicted')
            plt.title('Satisfaction Score Prediction')
            plt.xlabel('Time Step')
            plt.ylabel('Satisfaction Score')
            plt.legend()
            plt.tight_layout()
            plt.show()

            # Make future predictions
            # Get the last sequence
            last_sequence = satisfaction_scaled[-seq_length:].reshape(1, seq_length, 1)

            # Predict next 30 days
            forecast_steps = 30
            forecast = []

            current_sequence = last_sequence.copy()

            for _ in range(forecast_steps):
                # Predict next value
                next_pred = model.predict(current_sequence)[0][0]
                forecast.append(next_pred)

                # Update sequence
                current_sequence = np.append(current_sequence[0, 1:], [[next_pred]], axis=0)
                current_sequence = current_sequence.reshape(1, seq_length, 1)

            # Convert to original scale
            forecast = scaler.inverse_transform(np.array(forecast).reshape(-1, 1))

            # Create forecast dates
            last_date = daily_satisfaction.index[-1]
            forecast_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=forecast_steps)

            # Plot historical data and forecast
            plt.figure(figsize=(12, 6))
            plt.plot(daily_satisfaction.index[-60:], daily_satisfaction.values[-60:], label='Historical')
            plt.plot(forecast_dates, forecast, label='Forecast', color='red')
            plt.title('Satisfaction Score Forecast')
            plt.xlabel('Date')
            plt.ylabel('Satisfaction Score')
            plt.legend()
            plt.tight_layout()
            plt.show()

            # Save the model
            self.time_series_model = {
                'model': model,
                'scaler': scaler,
                'seq_length': seq_length,
                'forecast': forecast,
                'forecast_dates': forecast_dates
            }

            return self.time_series_model

        except Exception as e:
            print(f"Error building time series model: {e}")
            return None
# Main block (unchanged)
if __name__ == "__main__":
    print("Initializing Healthcare Management Models...")

    # Initialize models
    models = HealthcareManagementModels()

    # Run all models with sample data
    print("\nRunning all models with sample data...")
    results = models.run_all_models()  # Call run_all_models as a method

    # Save models
    if results:
        print("\nSaving all trained models...")
        saved_paths = models.save_models()
        print("\nSaved model paths:")
        for model_type, path in saved_paths.items():
            print(f"{model_type}: {path}")

        # Load models to verify
        print("\nLoading saved models to verify...")
        loaded_models = models.load_models()
        print("\nLoaded models status:")
        for model_type, model in loaded_models.items():
            print(f"{model_type}: {'Loaded' if model else 'Not loaded'}")

        # Generate insights
        print("\nGenerating insights from models...")
        insights = models.generate_insights()
        print("\nAnalysis complete!")
        print("\nInsights:")
        for model_type, insight in insights.items():
            print(f"\n{model_type.capitalize()}:")
            for key, value in insight.items():
                print(f"  {key}: {value}")
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load datasets
hospital_data = pd.read_csv('merged_hospital_admission_data.csv')
medical_data = pd.read_csv('merged_medical_care_data.csv')

print("Hospital Data Shape:", hospital_data.shape)
print("Medical Data Shape:", medical_data.shape)

# Check for missing values in key columns
print("\nMissing Values in Key Hospital Data Columns:")
key_columns = ['length_of_stay', 'total_cost', 'DIAGNOSIS', 'subject_id']
for col in key_columns:
    if col in hospital_data.columns:
        print(f"{col}: {hospital_data[col].isnull().sum()} missing values")

# Since length_of_stay is showing all missing values, let's fix this
# We'll need to synthesize this data for the demonstration

# First, let's see if we have any date columns to potentially calculate length of stay
date_columns = [col for col in hospital_data.columns if 'date' in col.lower()]
print("\nAvailable date columns:", date_columns)

# Let's generate synthetic length of stay values based on a realistic distribution
np.random.seed(42)  # For reproducibility
hospital_data['length_of_stay'] = np.random.lognormal(mean=1.5, sigma=0.6, size=len(hospital_data))
hospital_data['length_of_stay'] = hospital_data['length_of_stay'].round(1)  # Round to 1 decimal place

print("\nSynthetic Length of Stay Statistics:")
print(hospital_data['length_of_stay'].describe())

# Also add synthetic total_cost if it's missing
if hospital_data['total_cost'].isnull().all():
    # Generate costs that correlate with length of stay
    base_cost = 2000  # Base cost
    daily_cost = 1500  # Cost per day
    variation = 0.3    # Random variation factor

    hospital_data['total_cost'] = (
        base_cost +
        hospital_data['length_of_stay'] * daily_cost *
        np.random.uniform(1-variation, 1+variation, size=len(hospital_data))
    ).round(-2)  # Round to nearest 100

    print("\nSynthetic Total Cost Statistics:")
    print(hospital_data['total_cost'].describe())

# Create a properly formatted class with fixed data
class TreatmentDurationModels:
    def __init__(self):
        """Initialize the class with empty attributes."""
        self.hospital_data = hospital_data.copy()  # Using the fixed data
        self.medical_data = medical_data.copy()
        self.merged_data = None
        self.classification_model = None
        self.regression_model = None
        self.clustering_model = None
        self.time_series_model = None

    def preprocess_data(self):
        """Preprocess the data and create necessary features."""
        print("Preprocessing data...")

        # Create stay categories in hospital data
        q1 = self.hospital_data['length_of_stay'].quantile(0.33)
        q2 = self.hospital_data['length_of_stay'].quantile(0.67)

        self.hospital_data['stay_category'] = pd.cut(
            self.hospital_data['length_of_stay'],
            bins=[0, q1, q2, float('inf')],
            labels=['Short', 'Medium', 'Long']
        )

        print(f"Stay categories: Short <= {q1:.1f} days, Medium <= {q2:.1f} days, Long > {q2:.1f} days")

        # Convert BMI to numeric if it exists
        if 'BMI' in self.hospital_data.columns:
            self.hospital_data['BMI'] = pd.to_numeric(self.hospital_data['BMI'].str.replace(',', '.'), errors='coerce')

        # Create diagnosis category if present
        if 'DIAGNOSIS' in self.hospital_data.columns:
            # Extract main diagnosis category (first word)
            self.hospital_data['diagnosis_category'] = self.hospital_data['DIAGNOSIS'].apply(
                lambda x: str(x).split()[0] if pd.notna(x) else 'Unknown'
            )

        # Create comorbidity count if health condition columns are available
        health_condition_cols = [col for col in self.hospital_data.columns if col.startswith('Had')]
        if health_condition_cols:
            # Convert Yes/No to 1/0 for HadDiabetes if needed
            if 'HadDiabetes' in health_condition_cols and self.hospital_data['HadDiabetes'].dtype == 'object':
                self.hospital_data['HadDiabetes'] = self.hospital_data['HadDiabetes'].map({'Yes': 1, 'No': 0}).fillna(0)

            # Sum health conditions to create comorbidity count
            binary_health_cols = [col for col in health_condition_cols
                                 if self.hospital_data[col].dtype != 'object'
                                 or col == 'HadDiabetes']

            if binary_health_cols:
                self.hospital_data['comorbidity_count'] = self.hospital_data[binary_health_cols].sum(axis=1)

        # Fill missing values in hospital data
        # For categorical columns
        for col in self.hospital_data.select_dtypes(include=['object']).columns:
            self.hospital_data[col] = self.hospital_data[col].fillna('Unknown')

        # For numerical columns
        for col in self.hospital_data.select_dtypes(include=['number']).columns:
            if self.hospital_data[col].isnull().any():
                self.hospital_data[col] = self.hospital_data[col].fillna(self.hospital_data[col].median())

        # We'll use only the hospital data for modeling, so we don't need to merge
        # This avoids the mismatch in row counts
        self.merged_data = self.hospital_data.copy()

        print(f"Processed data shape: {self.merged_data.shape}")

        return True

# Create an instance and preprocess the data
models = TreatmentDurationModels()
models.preprocess_data()

# Verify the data is properly prepared
print("\nData Shape after preprocessing:", models.merged_data.shape)
print("Stay Category Distribution:")
print(models.merged_data['stay_category'].value_counts())

if 'comorbidity_count' in models.merged_data.columns:
    print("\nComorbidity Count Distribution:")
    print(models.merged_data['comorbidity_count'].value_counts().sort_index())

# Show correlations between key features
if models.merged_data is not None:
    numeric_cols = ['length_of_stay', 'total_cost']
    if 'BMI' in models.merged_data.columns:
        numeric_cols.append('BMI')
    if 'comorbidity_count' in models.merged_data.columns:
        numeric_cols.append('comorbidity_count')

    available_cols = [col for col in numeric_cols if col in models.merged_data.columns]

    if len(available_cols) >= 2:
        print("\nCorrelations between key features:")
        corr = models.merged_data[available_cols].corr()
        print(corr)

# Plot stay categories
plt.figure(figsize=(8, 6))
models.merged_data['stay_category'].value_counts().plot(kind='bar')
plt.title('Distribution of Length of Stay Categories')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

def build_classification_model(self):
    """
    Build a classification model to predict length of stay categories.
    Returns the trained model, preprocessor, and evaluation metrics.
    """
    print("\n=== Building Classification Model for Treatment Duration ===")

    # Define target
    target = 'stay_category'

    # Select categorical features
    cat_features = ['DIAGNOSIS', 'diagnosis_category', 'ADMISSION_TYPE', 'gender', 'ETHNICITY']
    cat_features = [f for f in cat_features if f in self.merged_data.columns]

    # Select numerical features
    num_features = []
    if 'BMI' in self.merged_data.columns and pd.to_numeric(self.merged_data['BMI'], errors='coerce').notna().any():
        num_features.append('BMI')
    if 'comorbidity_count' in self.merged_data.columns:
        num_features.append('comorbidity_count')
    if 'total_cost' in self.merged_data.columns:
        num_features.append('total_cost')

    print(f"Using categorical features: {cat_features}")
    print(f"Using numerical features: {num_features}")

    # Create preprocessing pipeline
    transformers = []
    if num_features:
        transformers.append(('num', StandardScaler(), num_features))
    if cat_features:
        transformers.append(('cat', OneHotEncoder(handle_unknown='ignore'), cat_features))

    preprocessor = ColumnTransformer(transformers=transformers)

    # Create model pipeline with SVM
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', SVC(probability=True, random_state=42))
    ])

    # Split data
    X = self.merged_data[cat_features + num_features]
    y = self.merged_data[target]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    print(f"Training data shape: {X_train.shape}")
    print(f"Testing data shape: {X_test.shape}")

    # Define parameter grid for optimization
    param_grid = {
        'classifier__C': [0.1, 1, 10],
        'classifier__gamma': ['scale', 'auto']
    }

    # Grid search for hyperparameter tuning
    grid_search = GridSearchCV(
        pipeline,
        param_grid,
        cv=5,
        scoring='accuracy',
        n_jobs=-1
    )

    # Train model
    grid_search.fit(X_train, y_train)

    print(f"Best parameters: {grid_search.best_params_}")

    # Get best model
    best_model = grid_search.best_estimator_

    # Evaluate model
    y_pred = best_model.predict(X_test)
    accuracy = (y_pred == y_test).mean()

    print(f"Model Accuracy: {accuracy:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=sorted(y.unique()), yticklabels=sorted(y.unique()))
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix - Length of Stay Classification')
    plt.tight_layout()
    plt.show()

    # Save the model
    self.classification_model = {
        'model': best_model,
        'features': cat_features + num_features,
        'accuracy': accuracy,
        'target': target
    }

    return self.classification_model

# Add method to class
TreatmentDurationModels.build_classification_model = build_classification_model

# Build classification model
classification_model = models.build_classification_model()
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

def build_regression_model(self):
    """
    Build a regression model to predict exact length of stay.
    Returns the trained model, preprocessor, and evaluation metrics.
    """
    print("\n=== Building Regression Model for Treatment Duration ===")

    # Define target
    target = 'length_of_stay'

    # Select categorical features (same as classification for consistency)
    cat_features = ['DIAGNOSIS', 'diagnosis_category', 'ADMISSION_TYPE', 'gender', 'ETHNICITY']
    cat_features = [f for f in cat_features if f in self.merged_data.columns]

    # Select numerical features - excluding total_cost as it may have high correlation with LOS
    num_features = []
    if 'BMI' in self.merged_data.columns and pd.to_numeric(self.merged_data['BMI'], errors='coerce').notna().any():
        num_features.append('BMI')
    if 'comorbidity_count' in self.merged_data.columns:
        num_features.append('comorbidity_count')

    print(f"Using categorical features: {cat_features}")
    print(f"Using numerical features: {num_features}")

    # Create preprocessing pipeline
    transformers = []
    if num_features:
        transformers.append(('num', StandardScaler(), num_features))
    if cat_features:
        transformers.append(('cat', OneHotEncoder(handle_unknown='ignore'), cat_features))

    preprocessor = ColumnTransformer(transformers=transformers)

    # Create model pipeline with Gradient Boosting Regressor
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('regressor', GradientBoostingRegressor(random_state=42))
    ])

    # Split data
    X = self.merged_data[cat_features + num_features]
    y = self.merged_data[target]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    print(f"Training data shape: {X_train.shape}")
    print(f"Testing data shape: {X_test.shape}")

    # Train model
    pipeline.fit(X_train, y_train)

    # Evaluate model
    y_pred = pipeline.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print(f"Mean Squared Error: {mse:.4f}")
    print(f"Root Mean Squared Error: {np.sqrt(mse):.4f}")
    print(f"R-squared: {r2:.4f}")

    # Feature importance
    if hasattr(pipeline['regressor'], 'feature_importances_'):
        # Get feature names after preprocessing
        feature_names = []

        # Get numeric feature names
        if num_features:
            feature_names.extend(num_features)

        # Get one-hot encoded feature names from categorical variables
        if cat_features:
            # Get all categories from one-hot encoder
            encoder = pipeline.named_steps['preprocessor'].transformers_[len(transformers)-1][1]
            try:
                # For sklearn >= 0.23
                cat_feature_names = encoder.get_feature_names_out(cat_features)
            except:
                # For sklearn < 0.23
                cat_feature_names = [f"{col}_{val}" for col in cat_features
                                    for val in self.merged_data[col].astype(str).unique()]

            feature_names.extend(cat_feature_names)

        # Map importances to feature names (take first n importances if needed)
        importances = pipeline['regressor'].feature_importances_
        n_features = min(len(importances), len(feature_names))

        feature_importance = pd.DataFrame({
            'Feature': feature_names[:n_features],
            'Importance': importances[:n_features]
        }).sort_values('Importance', ascending=False)

        print("\nTop 10 Important Features:")
        print(feature_importance.head(10))

        # Plot feature importance
        plt.figure(figsize=(12, 8))
        top_features = feature_importance.head(10)
        sns.barplot(x='Importance', y='Feature', data=top_features)
        plt.title('Feature Importance for Length of Stay Prediction')
        plt.tight_layout()
        plt.show()

    # Plot actual vs predicted
    plt.figure(figsize=(10, 6))
    plt.scatter(y_test, y_pred, alpha=0.5)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
    plt.xlabel('Actual Length of Stay')
    plt.ylabel('Predicted Length of Stay')
    plt.title('Actual vs Predicted Length of Stay')
    plt.tight_layout()
    plt.show()

    # Save the model
    self.regression_model = {
        'model': pipeline,
        'features': cat_features + num_features,
        'mse': mse,
        'r2': r2,
        'target': target
    }

    return self.regression_model

# Add method to class
TreatmentDurationModels.build_regression_model = build_regression_model

# Build regression model
regression_model = models.build_regression_model()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

def build_clustering_model(self):
    """
    Build a clustering model to identify unusual length of stay patterns.
    Returns the trained model and cluster analysis.
    """
    print("\n=== Building Clustering Model for Treatment Duration ===")

    # Select features for clustering
    cluster_features = ['length_of_stay']
    additional_features = ['total_cost', 'BMI', 'comorbidity_count']

    # Add available additional features
    cluster_features.extend([f for f in additional_features if f in self.merged_data.columns])

    # Check if we have the necessary features
    available_features = [f for f in cluster_features if f in self.merged_data.columns]

    if len(available_features) < 2:
        print("Error: Not enough features available for clustering")
        return None

    print(f"Using features for clustering: {available_features}")

    # Prepare data for clustering
    X = self.merged_data[available_features].copy()

    # Convert to numeric if needed
    for col in X.columns:
        if X[col].dtype == 'object':
            X[col] = pd.to_numeric(X[col], errors='coerce')

    # Drop rows with missing values
    X = X.dropna()

    if len(X) < 10:
        print("Error: Not enough complete data points for clustering")
        return None

    # Scale the features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Find optimal eps parameter using k-distance plot
    # Get k-distances
    k = 5  # number of neighbors
    nn = NearestNeighbors(n_neighbors=k)
    nn.fit(X_scaled)
    distances, indices = nn.kneighbors(X_scaled)

    # Sort distances to find elbow
    distance_desc = np.sort(distances[:, k-1])[::-1]

    # Plot k-distance graph
    plt.figure(figsize=(10, 6))
    plt.plot(distance_desc)
    plt.axhline(y=0.5, color='r', linestyle='--')
    plt.title('K-Distance Graph for DBSCAN Parameter Selection')
    plt.xlabel('Points sorted by distance')
    plt.ylabel(f'{k}-th nearest neighbor distance')
    plt.tight_layout()
    plt.show()

    # Select eps parameter (could be automated in production)
    eps = 0.5  # Example value, would be determined from k-distance plot
    min_samples = 5

    print(f"Selected DBSCAN parameters: eps={eps}, min_samples={min_samples}")

    # Build DBSCAN model to identify unusual patterns
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    clusters = dbscan.fit_predict(X_scaled)

    # Add cluster labels to original data
    X['cluster'] = clusters

    # Count number of clusters and noise points
    n_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
    n_noise = list(clusters).count(-1)

    print(f"Number of clusters: {n_clusters}")
    print(f"Number of noise points: {n_noise}")

    # Analyze clusters
    cluster_analysis = X.groupby('cluster').mean()
    print("\nCluster Analysis:")
    print(cluster_analysis)

    # Visualize clusters (first two features)
    plt.figure(figsize=(12, 8))

    if len(available_features) >= 2:
        scatter = plt.scatter(
            X_scaled[:, 0],
            X_scaled[:, 1],
            c=clusters,
            cmap='viridis',
            alpha=0.8
        )

        # Highlight noise points
        noise_points = (clusters == -1)
        plt.scatter(
            X_scaled[noise_points, 0],
            X_scaled[noise_points, 1],
            c='red',
            marker='x',
            s=100,
            label='Anomalies'
        )

        plt.colorbar(scatter, label='Cluster')
        plt.xlabel(available_features[0])
        plt.ylabel(available_features[1])
        plt.title('Treatment Duration Patterns')
        plt.legend()
        plt.tight_layout()
        plt.show()

    # Save the model
    self.clustering_model = {
        'model': dbscan,
        'scaler': scaler,
        'features': available_features,
        'cluster_analysis': cluster_analysis,
        'n_clusters': n_clusters,
        'n_noise': n_noise
    }

    return self.clustering_model

# Add method to class
TreatmentDurationModels.build_clustering_model = build_clustering_model

# Build clustering model
clustering_model = models.build_clustering_model()
```

```python
def build_time_series_model(self):
    """
    Build a time series model to forecast length of stay trends.
    Uses a simple forecasting approach that doesn't rely on Prophet.
    """
    print("\n=== Building Time Series Model for Treatment Duration ===")

    # Always use synthetic data for demonstration since we have date issues
    print("Creating synthetic time series data for demonstration.")

    # Create a date range for the past year with daily frequency
    today = pd.Timestamp.today()
    start_date = today - pd.Timedelta(days=365)
    dates = pd.date_range(start=start_date, end=today, freq='D')

    # Generate synthetic average length of stay for each day
    # Base value + weekly seasonality + slight upward trend + noise
    n_days = len(dates)
    baseline = 5.0  # Average LOS

    # Components
    trend = np.linspace(0, 0.5, n_days)  # Slight upward trend
    weekly = 0.7 * np.sin(np.linspace(0, 2 * n_days * np.pi / 7, n_days))  # Weekly pattern
    noise = np.random.normal(0, 0.3, n_days)  # Random noise

    # Combine components
    avg_los = baseline + trend + weekly + noise
    avg_los = np.maximum(avg_los, 1.0)  # Ensure no negative LOS

    # Create time series dataframe
    daily_los = pd.DataFrame({
        'ds': dates,
        'y': avg_los
    })

    print(f"Created synthetic time series with {len(daily_los)} data points")

    # Plot the time series
    plt.figure(figsize=(12, 6))
    plt.plot(daily_los['ds'], daily_los['y'])
    plt.title('Average Length of Stay Time Series (Synthetic)')
    plt.xlabel('Date')
    plt.ylabel('Length of Stay (days)')
    plt.tight_layout()
    plt.show()

    # Create a simple forecast using rolling average
    # Calculate 7-day rolling average
    rolling_avg = daily_los['y'].rolling(window=7, min_periods=1).mean()

    # Plot with rolling average
    plt.figure(figsize=(12, 6))
    plt.plot(daily_los['ds'], daily_los['y'], label='Daily Average')
    plt.plot(daily_los['ds'], rolling_avg, label='7-day Rolling Average', linewidth=2)
    plt.title('Average Length of Stay with Rolling Average')
    plt.xlabel('Date')
    plt.ylabel('Length of Stay (days)')
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Create a simple future forecast using last 30 days trend
    last_30 = daily_los.iloc[-30:]

    # Linear regression to get trend
    from sklearn.linear_model import LinearRegression
    X = np.arange(len(last_30)).reshape(-1, 1)
    y = last_30['y'].values
    model = LinearRegression().fit(X, y)

    # Project forward 90 days
    future_days = 90
    future_X = np.arange(len(last_30), len(last_30) + future_days).reshape(-1, 1)
    future_y = model.predict(future_X)

    # Create future dates
    last_date = last_30['ds'].iloc[-1]
    future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=future_days)

    # Plot with simple forecast
    plt.figure(figsize=(12, 6))
    plt.plot(daily_los['ds'], daily_los['y'], label='Historical')
    plt.plot(future_dates, future_y, label='Forecast (Next 90 Days)', linestyle='--')
    plt.title('Length of Stay Forecast')
    plt.xlabel('Date')
    plt.ylabel('Length of Stay (days)')
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Analyze seasonality with Fourier decomposition
    from scipy import signal

    # Calculate periodogram
    f, Pxx = signal.periodogram(daily_los['y'].values, fs=1)

    # Plot periodogram to find dominant frequencies
    plt.figure(figsize=(10, 6))
    plt.semilogy(f, Pxx)
    plt.title('Periodogram of Length of Stay')
    plt.xlabel('Frequency (1/day)')
    plt.ylabel('Power Spectral Density')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Create a seasonal decomposition if statsmodels is available
    try:
        from statsmodels.tsa.seasonal import seasonal_decompose

        # Decompose the time series
        result = seasonal_decompose(daily_los['y'], model='multiplicative', period=7)

        # Plot decomposition
        fig = plt.figure(figsize=(12, 10))
        fig = result.plot()
        plt.suptitle('Seasonal Decomposition of Length of Stay')
        plt.tight_layout()
        plt.show()
    except:
        print("statsmodels not available for seasonal decomposition")

    # Save the model
    self.time_series_model = {
        'model': model,
        'forecast': pd.DataFrame({
            'ds': pd.concat([daily_los['ds'], pd.Series(future_dates)]),
            'yhat': np.concatenate([daily_los['y'].values, future_y])
        }),
        'data': daily_los,
        'current_avg': daily_los['y'].mean(),
        'future_avg': future_y.mean()
    }

    return self.time_series_model

# Update the method in the class
TreatmentDurationModels.build_time_series_model = build_time_series_model

# Build time series model
time_series_model = models.build_time_series_model()
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(42)

# Step 1: Load the datasets
def load_datasets():
    """Load all healthcare datasets"""
    print("Loading healthcare datasets...")

    try:
        drug_data = pd.read_csv('merged_drug_consumption_data.csv')
        print(f"Loaded drug consumption data: {drug_data.shape[0]} rows, {drug_data.shape[1]} columns")
    except Exception as e:
        print(f"Could not load drug consumption data: {e}")
        drug_data = None

    try:
        healthcare_data = pd.read_csv('merged_healthcare_data.csv')
        print(f"Loaded healthcare data: {healthcare_data.shape[0]} rows, {healthcare_data.shape[1]} columns")
    except Exception as e:
        print(f"Could not load healthcare data: {e}")
        healthcare_data = None

    try:
        hospital_data = pd.read_csv('merged_hospital_admission_data.csv')
        print(f"Loaded hospital admission data: {hospital_data.shape[0]} rows, {hospital_data.shape[1]} columns")
    except Exception as e:
        print(f"Could not load hospital admission data: {e}")
        hospital_data = None

    try:
        medical_care_data = pd.read_csv('merged_medical_care_data.csv')
        print(f"Loaded medical care data: {medical_care_data.shape[0]} rows, {medical_care_data.shape[1]} columns")
    except Exception as e:
        print(f"Could not load medical care data: {e}")
        medical_care_data = None

    return {
        'drug_data': drug_data,
        'healthcare_data': healthcare_data,
        'hospital_data': hospital_data,
        'medical_care_data': medical_care_data
    }

# Step 2: Data preprocessing functions
def preprocess_hospital_data(hospital_data):
    """Preprocess hospital admission data"""
    print("\nPreprocessing hospital admission data...")

    processed_data = hospital_data.copy()

    # Convert date columns to datetime
    date_cols = [col for col in processed_data.columns if 'Date_' in col]
    for col in date_cols:
        try:
            processed_data[col] = pd.to_datetime(processed_data[col], errors='coerce')
        except:
            print(f"Could not convert {col} to datetime")

    # Calculate length of stay from admission and discharge dates
    if 'Date_admission' in processed_data.columns and 'Date_discharge' in processed_data.columns:
        processed_data['length_of_stay'] = (
            processed_data['Date_discharge'] - processed_data['Date_admission']
        ).dt.total_seconds() / (24 * 60 * 60)  # Convert to days
        print(f"Calculated length of stay for {processed_data['length_of_stay'].notna().sum()} patients")

    # Calculate patient age from DOB and admission date
    if 'dob' in processed_data.columns and 'Date_admission' in processed_data.columns:
        try:
            processed_data['dob'] = pd.to_datetime(processed_data['dob'], errors='coerce')
            processed_data['age'] = (
                (processed_data['Date_admission'] - processed_data['dob']).dt.total_seconds()
                / (365.25 * 24 * 60 * 60)  # Convert to years
            )
            # Filter out unrealistic ages
            processed_data.loc[(processed_data['age'] < 0) | (processed_data['age'] > 120), 'age'] = np.nan
            print(f"Calculated age for {processed_data['age'].notna().sum()} patients")
        except:
            print("Could not calculate patient age")

    # Create binary features from categorical variables
    if 'gender' in processed_data.columns:
        processed_data['is_male'] = (processed_data['gender'] == 'M').astype(int)

    if 'ADMISSION_TYPE' in processed_data.columns:
        processed_data['is_emergency'] = (processed_data['ADMISSION_TYPE'] == 'EMERGENCY').astype(int)
        processed_data['is_elective'] = (processed_data['ADMISSION_TYPE'] == 'ELECTIVE').astype(int)
        processed_data['is_urgent'] = (processed_data['ADMISSION_TYPE'] == 'URGENT').astype(int)

    # Convert numeric columns from string to float
    numeric_cols = ['BMI', 'WeightInKilograms', 'HeightInMeters']
    for col in numeric_cols:
        if col in processed_data.columns:
            processed_data[col] = pd.to_numeric(processed_data[col], errors='coerce')

    # Handle medical history columns
    medical_cols = [col for col in processed_data.columns if col.startswith('Had')]
    for col in medical_cols:
        if col in processed_data.columns:
            if col == 'HadDiabetes':
                processed_data[col] = (processed_data[col] == 'Yes').astype(int)
            else:
                processed_data[col] = pd.to_numeric(processed_data[col], errors='coerce')

    print(f"Preprocessed hospital data: {processed_data.shape[0]} rows, {processed_data.shape[1]} columns")
    return processed_data

def preprocess_drug_data(drug_data):
    """Preprocess drug consumption data"""
    print("\nPreprocessing drug consumption data...")

    processed_data = drug_data.copy()

    # Convert date columns to datetime
    date_cols = [col for col in processed_data.columns if 'Date_' in col]
    for col in date_cols:
        try:
            processed_data[col] = pd.to_datetime(processed_data[col], errors='coerce')
        except:
            print(f"Could not convert {col} to datetime")

    # Aggregate drug consumption by patient
    print("Aggregating drug data by patient...")
    patient_drug_data = processed_data.groupby('subject_id').agg(
        total_drugs=('drug', 'count'),
        unique_drugs=('drug', 'nunique')
    ).reset_index()

    # Count drugs by type
    drug_types = ['MAIN', 'BASE', 'ADDITIVE']
    for drug_type in drug_types:
        type_count = processed_data[processed_data['drug_type'] == drug_type].groupby('subject_id').size()
        patient_drug_data[f'{drug_type.lower()}_drugs'] = patient_drug_data['subject_id'].map(type_count).fillna(0)

    # Count drugs by route
    common_routes = ['IV', 'PO', 'SC', 'IM']
    for route in common_routes:
        route_count = processed_data[processed_data['route'] == route].groupby('subject_id').size()
        patient_drug_data[f'{route.lower()}_drugs'] = patient_drug_data['subject_id'].map(route_count).fillna(0)

    # Count weekend vs weekday drugs
    if 'IsWeekend_start' in processed_data.columns:
        weekend_count = processed_data[processed_data['IsWeekend_start'] == True].groupby('subject_id').size()
        weekday_count = processed_data[processed_data['IsWeekend_start'] == False].groupby('subject_id').size()

        patient_drug_data['weekend_drugs'] = patient_drug_data['subject_id'].map(weekend_count).fillna(0)
        patient_drug_data['weekday_drugs'] = patient_drug_data['subject_id'].map(weekday_count).fillna(0)

        # Calculate weekend ratio
        total_drugs = patient_drug_data['weekend_drugs'] + patient_drug_data['weekday_drugs']
        patient_drug_data['weekend_ratio'] = (
            patient_drug_data['weekend_drugs'] / total_drugs
        ).fillna(0)

    print(f"Created aggregated drug data: {patient_drug_data.shape[0]} patients, {patient_drug_data.shape[1]} features")
    return patient_drug_data

def preprocess_healthcare_data(healthcare_data):
    """Preprocess healthcare data (complaints/satisfaction)"""
    print("\nPreprocessing healthcare data...")

    processed_data = healthcare_data.copy()

    # Convert date fields
    if 'Date' in processed_data.columns:
        processed_data['Date'] = pd.to_datetime(processed_data['Date'], errors='coerce')

    # Process satisfaction data
    if 'Resolution_Duration_Days' in processed_data.columns:
        processed_data['Resolution_Duration_Days'] = pd.to_numeric(
            processed_data['Resolution_Duration_Days'], errors='coerce')

    # Create binary features for complaint types
    if 'Complaint_Type' in processed_data.columns:
        complaint_types = processed_data['Complaint_Type'].unique()
        for complaint_type in complaint_types:
            processed_data[f'is_{complaint_type.lower()}_complaint'] = (
                processed_data['Complaint_Type'] == complaint_type).astype(int)

    print(f"Preprocessed healthcare data: {processed_data.shape[0]} rows, {processed_data.shape[1]} columns")
    return processed_data

def preprocess_medical_care_data(medical_care_data):
    """Preprocess medical care data"""
    print("\nPreprocessing medical care data...")

    processed_data = medical_care_data.copy()

    # Convert date fields
    if 'Date' in processed_data.columns:
        processed_data['Date'] = pd.to_datetime(processed_data['Date'], errors='coerce')

    # Aggregate treatments by patient and category
    if 'CATEGORY' in processed_data.columns and 'subject_id' in processed_data.columns:
        print("Aggregating medical care data by patient and category...")
        category_pivot = processed_data.pivot_table(
            index='subject_id',
            columns='CATEGORY',
            values='seq_num_treatment',
            aggfunc='count',
            fill_value=0
        ).reset_index()

        # Rename columns to avoid spaces and special characters
        category_pivot.columns = ['subject_id'] + [
            f'treatment_{col.lower().replace(" ", "_")}'
            for col in category_pivot.columns if col != 'subject_id'
        ]

        print(f"Created treatment category data: {category_pivot.shape[0]} patients, {category_pivot.shape[1]} features")
        return category_pivot

    print(f"Preprocessed medical care data: {processed_data.shape[0]} rows, {processed_data.shape[1]} columns")
    return processed_data

# Step 3: Exploratory Data Analysis
def exploratory_data_analysis(datasets):
    """Perform basic exploratory analysis on all datasets"""
    print("\nPerforming exploratory data analysis...")

    results = {}

    for name, dataset in datasets.items():
        if dataset is None:
            continue

        print(f"\nAnalyzing {name}:")

        # Basic dataset info
        print(f"Shape: {dataset.shape}")
        print(f"Columns: {dataset.columns.tolist()[:5]}...")

        # Check data types
        print("\nData types:")
        print(dataset.dtypes.value_counts())

        # Check for missing values
        missing_vals = dataset.isnull().sum()
        print("\nColumns with missing values:")
        print(missing_vals[missing_vals > 0].sort_values(ascending=False).head())

        # For numeric columns, show basic statistics
        numeric_cols = dataset.select_dtypes(include=[np.number]).columns
        if len(numeric_cols) > 0:
            print("\nStatistics for numeric columns:")
            stats = dataset[numeric_cols].describe().T
            print(stats[['count', 'mean', 'min', 'max']].head())

        # For categorical columns, show value counts
        categorical_cols = dataset.select_dtypes(include=['object']).columns
        if len(categorical_cols) > 0:
            print("\nTop categories for selected categorical columns:")
            for col in categorical_cols[:3]:  # Show first 3 categorical columns
                print(f"\n{col}:")
                print(dataset[col].value_counts().head(3))

        results[name] = {
            'shape': dataset.shape,
            'missing_values': missing_vals[missing_vals > 0].to_dict(),
            'numeric_columns': numeric_cols.tolist(),
            'categorical_columns': categorical_cols.tolist()
        }

    return results

# Main function to run all processing steps
def process_all_data():
    """Load and process all datasets"""
    # Load the data
    datasets = load_datasets()

    # Process each dataset
    processed_datasets = {}

    if datasets['hospital_data'] is not None:
        processed_datasets['hospital_data'] = preprocess_hospital_data(datasets['hospital_data'])

    if datasets['drug_data'] is not None:
        processed_datasets['drug_data'] = preprocess_drug_data(datasets['drug_data'])

    if datasets['healthcare_data'] is not None:
        processed_datasets['healthcare_data'] = preprocess_healthcare_data(datasets['healthcare_data'])

    if datasets['medical_care_data'] is not None:
        processed_datasets['medical_care_data'] = preprocess_medical_care_data(datasets['medical_care_data'])

    # Perform exploratory data analysis
    eda_results = exploratory_data_analysis(processed_datasets)

    return processed_datasets, eda_results

# Run the data processing
if __name__ == "__main__":
    processed_data, eda_results = process_all_data()
    print("\nData processing complete!")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error, silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.holtwinters import ExponentialSmoothing, SimpleExpSmoothing
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(42)

# Assuming you've already run the data processing code and have these variables:
# processed_data containing hospital_data, drug_data, healthcare_data, medical_care_data

def build_resource_allocation_models(hospital_data, drug_data):
    """Build models for resource allocation optimization"""
    print("\n===== OBJECTIVE 1: STREAMLINE RESOURCE ALLOCATION =====")

    # Merge datasets on subject_id
    print("Merging hospital and drug data...")
    resource_data = pd.merge(hospital_data, drug_data, on='subject_id', how='left')
    print(f"Merged data shape: {resource_data.shape}")

    # Fill missing drug data with zeros
    drug_cols = [col for col in drug_data.columns if col != 'subject_id']
    resource_data[drug_cols] = resource_data[drug_cols].fillna(0)

    # Check if length_of_stay is available
    if 'length_of_stay' not in resource_data.columns or resource_data['length_of_stay'].isna().all():
        print("Length of stay data not available. Generating synthetic LOS data for demonstration.")

        # Generate synthetic LOS based on patient and admission characteristics
        np.random.seed(42)  # For reproducibility

        # Base LOS - typically 3-7 days for most admissions
        resource_data['length_of_stay'] = np.random.normal(5, 2, size=len(resource_data))

        # Adjust based on admission type
        if 'is_emergency' in resource_data.columns:
            # Emergency admissions tend to have longer stays
            resource_data.loc[resource_data['is_emergency'] == 1, 'length_of_stay'] += 2

        if 'is_elective' in resource_data.columns:
            # Elective admissions tend to have shorter stays
            resource_data.loc[resource_data['is_elective'] == 1, 'length_of_stay'] -= 1

        # Adjust based on diagnosis (if available)
        if 'DIAGNOSIS' in resource_data.columns:
            # Some diagnoses have longer LOS
            longer_los_diagnoses = ['SEPSIS', 'PNEUMONIA', 'RESPIRATORY FAILURE', 'STROKE', 'HEART FAILURE']
            mask = resource_data['DIAGNOSIS'].isin(longer_los_diagnoses)
            resource_data.loc[mask, 'length_of_stay'] += 3

        # Ensure positive LOS
        resource_data['length_of_stay'] = np.maximum(1, resource_data['length_of_stay'])

        print(f"Generated synthetic length of stay for {len(resource_data)} patients")

    # Calculate target for classification: above or below median LOS
    median_los = resource_data['length_of_stay'].median()
    resource_data['los_above_median'] = (resource_data['length_of_stay'] > median_los).astype(int)

    print(f"Median length of stay: {median_los:.2f} days")
    print(f"Patients with above-median LOS: {resource_data['los_above_median'].sum()} ({resource_data['los_above_median'].mean()*100:.1f}%)")

    # Define features for modeling
    features = [
        'is_emergency', 'is_elective', 'is_urgent',
        'total_drugs', 'main_drugs', 'base_drugs',
        'iv_drugs', 'po_drugs', 'weekend_ratio'
    ]

    # Add medical history features if available
    medical_features = ['HadHeartAttack', 'HadDiabetes', 'HadAsthma', 'HadDepressiveDisorder']
    for feature in medical_features:
        if feature in resource_data.columns and not resource_data[feature].isna().all():
            features.append(feature)

    # Add age if available
    if 'age' in resource_data.columns and not resource_data['age'].isna().all():
        features.append('age')

    # Add gender if available
    if 'is_male' in resource_data.columns:
        features.append('is_male')

    print(f"Selected {len(features)} features for modeling: {features}")

    # Ensure all selected features are in the dataframe
    for feature in list(features):
        if feature not in resource_data.columns:
            print(f"Warning: Feature {feature} not found in data, removing it")
            features.remove(feature)

    # Prepare data for modeling
    model_data = resource_data.dropna(subset=['length_of_stay'])

    # Handle missing values in features with mean imputation
    for feature in features:
        if model_data[feature].isna().any():
            mean_val = model_data[feature].mean()
            model_data[feature] = model_data[feature].fillna(mean_val)

    print(f"Complete data for modeling: {len(model_data)} rows")

    # Split data for training and testing
    X = model_data[features]
    y_class = model_data['los_above_median']
    y_reg = model_data['length_of_stay']

    X_train, X_test, y_class_train, y_class_test, y_reg_train, y_reg_test = train_test_split(
        X, y_class, y_reg, test_size=0.2, random_state=42)

    print(f"Training data: {X_train.shape[0]} rows")
    print(f"Testing data: {X_test.shape[0]} rows")

    # Scale features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # 1. SUPERVISED CLASSIFICATION MODEL - Predict if LOS is above median
    print("\n----- 1. Classification Model for Length of Stay -----")

    # Try both Logistic Regression and Random Forest
    models = {
        'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
        'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42)
    }

    best_accuracy = 0
    best_model_name = None
    best_clf = None

    for name, model in models.items():
        model.fit(X_train_scaled, y_class_train)
        y_pred = model.predict(X_test_scaled)
        accuracy = accuracy_score(y_class_test, y_pred)
        print(f"{name} Accuracy: {accuracy:.4f}")

        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_model_name = name
            best_clf = model

    print(f"\nBest classification model: {best_model_name} (Accuracy: {best_accuracy:.4f})")

    # Feature importance for the best model
    if best_model_name == 'Logistic Regression':
        importance = pd.DataFrame({
            'feature': features,
            'importance': np.abs(best_clf.coef_[0])
        }).sort_values(by='importance', ascending=False)
    else:  # Random Forest
        importance = pd.DataFrame({
            'feature': features,
            'importance': best_clf.feature_importances_
        }).sort_values(by='importance', ascending=False)

    print("\nTop 5 features for predicting extended LOS:")
    print(importance.head(5))

    # Visualize feature importance
    plt.figure(figsize=(10, 6))
    sns.barplot(x='importance', y='feature', data=importance.head(10))
    plt.title(f'Top 10 Features for LOS Classification ({best_model_name})')
    plt.tight_layout()
    plt.show()

    # 2. SUPERVISED REGRESSION MODEL - Predict exact LOS
    print("\n----- 2. Regression Model for Length of Stay -----")

    # Try both Linear Regression and Random Forest
    reg_models = {
        'Linear Regression': LinearRegression(),
        'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42)
    }

    best_mae = float('inf')
    best_reg_name = None
    best_reg = None

    for name, model in reg_models.items():
        model.fit(X_train_scaled, y_reg_train)
        y_pred = model.predict(X_test_scaled)
        mse = mean_squared_error(y_reg_test, y_pred)
        mae = mean_absolute_error(y_reg_test, y_pred)
        print(f"{name} - MSE: {mse:.4f}, MAE: {mae:.4f} days")

        if mae < best_mae:
            best_mae = mae
            best_reg_name = name
            best_reg = model

    print(f"\nBest regression model: {best_reg_name} (MAE: {best_mae:.4f} days)")

    # Feature importance for the best regression model
    if best_reg_name == 'Linear Regression':
        reg_importance = pd.DataFrame({
            'feature': features,
            'importance': np.abs(best_reg.coef_)
        }).sort_values(by='importance', ascending=False)
    else:  # Random Forest
        reg_importance = pd.DataFrame({
            'feature': features,
            'importance': best_reg.feature_importances_
        }).sort_values(by='importance', ascending=False)

    print("\nTop 5 features for predicting exact LOS:")
    print(reg_importance.head(5))

    # Visualize regression feature importance
    plt.figure(figsize=(10, 6))
    sns.barplot(x='importance', y='feature', data=reg_importance.head(10))
    plt.title(f'Top 10 Features for LOS Prediction ({best_reg_name})')
    plt.tight_layout()
    plt.show()

    # 3. UNSUPERVISED LEARNING - Cluster patients by resource usage
    print("\n----- 3. Clustering Model for Resource Allocation -----")

    # Select features for clustering
    cluster_features = ['length_of_stay', 'total_drugs']

    # Add other important features based on previous models
    top_features = list(importance['feature'].head(3))
    for feature in top_features:
        if feature not in cluster_features:
            cluster_features.append(feature)

    # Keep it to max 5-6 features for better interpretability
    cluster_features = cluster_features[:6]
    print(f"Selected {len(cluster_features)} features for clustering: {cluster_features}")

    # Prepare data for clustering
    cluster_data = model_data[cluster_features].copy()

    # Scale data
    cluster_scaler = StandardScaler()
    cluster_data_scaled = cluster_scaler.fit_transform(cluster_data)

    # Determine optimal number of clusters using silhouette score
    silhouette_scores = []
    max_clusters = min(6, len(cluster_data) // 5)  # Don't try too many clusters

    for k in range(2, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        cluster_labels = kmeans.fit_predict(cluster_data_scaled)

        # Only calculate silhouette if we have enough samples
        if len(set(cluster_labels)) > 1:  # Need at least 2 clusters
            score = silhouette_score(cluster_data_scaled, cluster_labels)
            silhouette_scores.append((k, score))
            print(f"K={k}, Silhouette Score: {score:.4f}")

    # Choose best k
    if silhouette_scores:
        best_k = max(silhouette_scores, key=lambda x: x[1])[0]
    else:
        best_k = 3  # Default if we can't calculate silhouette

    print(f"Selected optimal number of clusters: {best_k}")

    # Final clustering
    kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
    model_data['resource_cluster'] = kmeans.fit_predict(cluster_data_scaled)

    # Analyze clusters
    cluster_analysis = model_data.groupby('resource_cluster').agg({
        'length_of_stay': 'mean',
        'total_drugs': 'mean',
        'is_emergency': 'mean' if 'is_emergency' in model_data.columns else None,
        'subject_id': 'count'
    }).rename(columns={'subject_id': 'count'})

    # Remove None columns
    cluster_analysis = cluster_analysis.loc[:, cluster_analysis.columns.notnull()]

    print("\nCluster analysis:")
    print(cluster_analysis)

    # Visualize clusters
    if len(cluster_features) >= 2:
        plt.figure(figsize=(10, 8))

        # Use first two features for visualization
        feature1, feature2 = cluster_features[0], cluster_features[1]

        # Plot the clusters
        scatter = plt.scatter(
            model_data[feature1],
            model_data[feature2],
            c=model_data['resource_cluster'],
            cmap='viridis',
            s=50,
            alpha=0.8
        )

        plt.colorbar(scatter, label='Cluster')
        plt.xlabel(feature1)
        plt.ylabel(feature2)
        plt.title(f'Resource Allocation Clusters (k={best_k})')
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()

    # 4. TIME SERIES MODEL - Admissions prediction
    print("\n----- 4. Time Series Model for Resource Allocation -----")

    # Create time series data - daily admissions
    if 'Date_admission' in hospital_data.columns:
        # Try to use real dates
        hospital_data['Date_admission'] = pd.to_datetime(hospital_data['Date_admission'])
        daily_admissions = hospital_data.groupby(hospital_data['Date_admission'].dt.date).size().reset_index()
        daily_admissions.columns = ['date', 'admissions']
    else:
        print("Creating synthetic time series data for demonstration")

        # Generate 60 days of synthetic admission data
        np.random.seed(42)
        start_date = pd.Timestamp('2023-01-01')
        dates = [start_date + pd.Timedelta(days=i) for i in range(60)]

        # Create cyclical pattern (weekly)
        base_admissions = 10 + np.sin(np.arange(60) * 2 * np.pi / 7) * 5

        # Add random noise
        noise = np.random.normal(0, 2, 60)
        admissions = np.maximum(1, base_admissions + noise).astype(int)

        daily_admissions = pd.DataFrame({
            'date': dates,
            'admissions': admissions
        })

    # Sort by date
    daily_admissions = daily_admissions.sort_values('date')
    print(f"Time series data: {len(daily_admissions)} days")

    # Plot time series
    plt.figure(figsize=(12, 6))
    plt.plot(daily_admissions['date'], daily_admissions['admissions'], marker='o', linestyle='-')
    plt.title('Daily Hospital Admissions')
    plt.ylabel('Number of Admissions')
    plt.xlabel('Date')
    plt.xticks(rotation=45)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    # Split into train and test
    train_size = int(len(daily_admissions) * 0.8)
    train = daily_admissions[:train_size]
    test = daily_admissions[train_size:]

    # Simple forecasting models
    if len(train) >= 7:  # Need at least a week of data
        try:
            # Moving average model (7-day)
            train['admissions_ma7'] = train['admissions'].rolling(window=7).mean()

            # Simple exponential smoothing
            model = SimpleExpSmoothing(train['admissions'].dropna())
            model_fit = model.fit(smoothing_level=0.3)
            train['admissions_ses'] = model_fit.fittedvalues

            # Forecast for test period
            ses_forecast = model_fit.forecast(len(test))

            # Calculate errors
            test = test.copy()  # Avoid SettingWithCopyWarning
            test['forecast_ses'] = ses_forecast.values
            mae = mean_absolute_error(test['admissions'], test['forecast_ses'])
            mse = mean_squared_error(test['admissions'], test['forecast_ses'])

            print(f"\nTime series forecasting results:")
            print(f"MAE: {mae:.4f}")
            print(f"MSE: {mse:.4f}")

            # Plot forecasts
            plt.figure(figsize=(12, 6))
            plt.plot(train['date'], train['admissions'], label='Training Data', marker='o', alpha=0.7)
            plt.plot(test['date'], test['admissions'], label='Actual Test Data', marker='o', alpha=0.7)
            plt.plot(test['date'], test['forecast_ses'], label='SES Forecast', marker='s', linestyle='--')
            plt.title('Hospital Admissions Forecast')
            plt.legend()
            plt.ylabel('Number of Admissions')
            plt.xlabel('Date')
            plt.xticks(rotation=45)
            plt.grid(True, linestyle='--', alpha=0.5)
            plt.tight_layout()
            plt.show()

        except Exception as e:
            print(f"Error in time series modeling: {e}")
    else:
        print("Insufficient data for time series modeling")

    # Return the models and results
    return {
        'classification_model': best_clf,
        'classification_accuracy': best_accuracy,
        'regression_model': best_reg,
        'regression_mae': best_mae,
        'clustering_model': kmeans,
        'clusters': cluster_analysis,
        'feature_importance': importance,
        'reg_feature_importance': reg_importance
    }

# Load the processed data from the previous step
# If you saved the processed data from the previous step, load it here:
# hospital_data = pd.read_csv('processed_hospital_data.csv')
# drug_data = pd.read_csv('processed_drug_data.csv')

# Otherwise, use the variables from the previous code execution:
# Assuming processed_data is a dictionary with keys 'hospital_data', 'drug_data', etc.

# Now build the resource allocation models
resource_models = build_resource_allocation_models(
    processed_data['hospital_data'],
    processed_data['drug_data']
)

print("\nResource allocation modeling complete!")
```

```python
# Data preprocessing for complication prediction

# Define our target variable - whether a patient had complications
hospital_data['had_complications'] = 0

# Mark complications based on discharge location
complication_locations = ['DEAD/EXPIRED', 'ICU', 'SNF']
hospital_data.loc[hospital_data['DISCHARGE_LOCATION'].isin(complication_locations), 'had_complications'] = 1

# Mark complications based on diagnosis
complication_diagnoses = ['SEPSIS', 'FAILURE', 'RESPIRATORY DISTRESS', 'COMPLICATION', 'PNEUMONIA']
for diagnosis in complication_diagnoses:
    hospital_data.loc[hospital_data['DIAGNOSIS'].str.contains(diagnosis, case=False, na=False), 'had_complications'] = 1

# Instead of calculating exact age, let's just use the year difference
# Extract year from dob and admission date
hospital_data['birth_year'] = pd.to_datetime(hospital_data['dob']).dt.year
hospital_data['admission_year'] = pd.to_datetime(hospital_data['Date_admission']).dt.year
hospital_data['age_approx'] = hospital_data['admission_year'] - hospital_data['birth_year']

# Fix numeric columns with comma decimal separator
for col in ['HeightInMeters', 'WeightInKilograms', 'BMI']:
    hospital_data[col] = hospital_data[col].str.replace(',', '.').astype(float)

# Convert diabetes from string to numeric
hospital_data['has_diabetes'] = hospital_data['HadDiabetes'].apply(
    lambda x: 1 if str(x).startswith('Yes') else 0
)

# Check complication distribution
print("Complication distribution:")
print(hospital_data['had_complications'].value_counts())
print(f"Percentage of admissions with complications: {hospital_data['had_complications'].mean() * 100:.2f}%")

# Display age distribution
print("\nAge approximation statistics:")
print(hospital_data['age_approx'].describe())
```

```python
# Model 1: Supervised Classification Model for Complication Prediction
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, roc_curve, confusion_matrix
from sklearn.model_selection import cross_val_score, StratifiedKFold
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create full pipeline with preprocessing and model
clf = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42))
])

# Train the model on the full training set
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)
y_pred_prob = clf.predict_proba(X_test)[:, 1]

# Evaluate the model
print("\nClassification Model Performance:")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"ROC AUC: {roc_auc_score(y_test, y_pred_prob):.4f}")

print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
print(cm)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

# Plot ROC curve
plt.figure(figsize=(8, 6))
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc_score(y_test, y_pred_prob):.4f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Complication Prediction')
plt.legend()
plt.show()

# Try to get feature importance
try:
    # Get preprocessed feature names
    all_features = []

    # Add numerical features
    all_features.extend(numerical_features)

    # Add one-hot encoded features
    ohe = clf.named_steps['preprocessor'].named_transformers_['cat'].named_steps['onehot']
    categorical_features_encoded = ohe.get_feature_names_out(categorical_features)
    all_features.extend(categorical_features_encoded)

    # Get feature importances and sort
    importances = clf[-1].feature_importances_
    indices = np.argsort(importances)[-15:]  # Top 15 features

    plt.figure(figsize=(10, 8))
    plt.title('Top 15 Feature Importances for Complication Prediction')
    plt.barh(range(len(indices)), importances[indices], align='center')
    plt.yticks(range(len(indices)), [all_features[i] for i in indices])
    plt.xlabel('Relative Importance')
    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"Could not generate feature importance plot: {e}")
```

```python
# Model 3: Unsupervised Clustering Model
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA

# Select numerical features for clustering
numerical_cluster_features = [
    'age_approx', 'HeightInMeters', 'WeightInKilograms', 'BMI',
    'HadHeartAttack', 'HadAngina', 'HadStroke', 'HadAsthma',
    'HadSkinCancer', 'HadCOPD', 'HadDepressiveDisorder',
    'HadKidneyDisease', 'HadArthritis', 'has_diabetes'
]

# Prepare data for clustering
X_cluster = hospital_data[numerical_cluster_features].copy()

# Impute missing values
imputer = SimpleImputer(strategy='median')
X_cluster_imputed = imputer.fit_transform(X_cluster)

# Scale the data
scaler = StandardScaler()
X_cluster_scaled = scaler.fit_transform(X_cluster_imputed)

# Find optimal number of clusters
silhouette_scores = []
k_range = range(2, 6)  # Using a smaller range to save time

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_cluster_scaled)
    score = silhouette_score(X_cluster_scaled, cluster_labels)
    silhouette_scores.append(score)
    print(f"K = {k}, Silhouette Score = {score:.4f}")

# Plot silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different k Values')
plt.grid(True)
plt.show()

# Get optimal k
optimal_k = k_range[np.argmax(silhouette_scores)]
print(f"Optimal number of clusters: {optimal_k}")

# Train final clustering model with optimal k
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
hospital_data['cluster'] = kmeans.fit_predict(X_cluster_scaled)

# Analyze complication rates by cluster
cluster_comp_rates = hospital_data.groupby('cluster')['had_complications'].mean() * 100
print("\nComplication rates by cluster:")
print(cluster_comp_rates)

# Apply PCA to reduce to 2 dimensions for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_cluster_scaled)

# Add PCA components to DataFrame
hospital_data['pca_1'] = X_pca[:, 0]
hospital_data['pca_2'] = X_pca[:, 1]

# Plot the clusters
plt.figure(figsize=(12, 8))
for cluster in range(optimal_k):
    subset = hospital_data[hospital_data['cluster'] == cluster]
    plt.scatter(subset['pca_1'], subset['pca_2'], label=f'Cluster {cluster}')

plt.title('Patient Clusters (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()

# Plot clusters with complication status
plt.figure(figsize=(12, 8))
colors = {0: 'blue', 1: 'red'}
complications = {0: 'No Complications', 1: 'Complications'}

for comp_status in [0, 1]:
    subset = hospital_data[hospital_data['had_complications'] == comp_status]
    plt.scatter(subset['pca_1'], subset['pca_2'],
                c=colors[comp_status],
                label=complications[comp_status],
                alpha=0.7)

plt.title('Patient Clusters by Complication Status (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Model 4: Time Series Model
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Let's create a time series of complication rates over time
# First, we'll extract admission dates and complication status
ts_data = hospital_data[['Date_admission', 'had_complications']].copy()
ts_data['Date_admission'] = pd.to_datetime(ts_data['Date_admission'])

# Group by admission date and calculate daily complication rate
daily_comp_rate = ts_data.groupby('Date_admission')['had_complications'].mean().reset_index()
daily_comp_rate = daily_comp_rate.set_index('Date_admission')
daily_comp_rate = daily_comp_rate.sort_index()

# Fill missing dates with forward filling (or you could use a rolling average)
date_range = pd.date_range(start=daily_comp_rate.index.min(), end=daily_comp_rate.index.max())
daily_comp_rate = daily_comp_rate.reindex(date_range).fillna(method='ffill')

# Plot the time series of complication rates
plt.figure(figsize=(14, 7))
plt.plot(daily_comp_rate.index, daily_comp_rate['had_complications'])
plt.title('Daily Complication Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Complication Rate')
plt.grid(True)
plt.tight_layout()
plt.show()

# Calculate a 7-day moving average for smoother trend
daily_comp_rate['7d_ma'] = daily_comp_rate['had_complications'].rolling(window=7, min_periods=1).mean()

plt.figure(figsize=(14, 7))
plt.plot(daily_comp_rate.index, daily_comp_rate['had_complications'], label='Daily Rate', alpha=0.5)
plt.plot(daily_comp_rate.index, daily_comp_rate['7d_ma'], label='7-Day Moving Average', color='red')
plt.title('Complication Rate with 7-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Complication Rate')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Try to fit a simple ARIMA model
try:
    # First check if we have enough data
    print(f"Number of time series data points: {len(daily_comp_rate)}")

    if len(daily_comp_rate) > 30:  # Need sufficient data for ARIMA
        # Try a simple model
        model = ARIMA(daily_comp_rate['7d_ma'], order=(1, 1, 1))
        model_fit = model.fit()

        # Forecast for the next 14 days
        forecast_steps = 14
        forecast = model_fit.forecast(steps=forecast_steps)

        # Create forecast dates
        last_date = daily_comp_rate.index[-1]
        forecast_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=forecast_steps)

        # Plot the forecast
        plt.figure(figsize=(14, 7))
        plt.plot(daily_comp_rate.index, daily_comp_rate['7d_ma'], label='Historical Data')
        plt.plot(forecast_dates, forecast, label='Forecast', color='red')
        plt.title('Complication Rate Forecast')
        plt.xlabel('Date')
        plt.ylabel('Complication Rate')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    else:
        print("Not enough data points for ARIMA modeling")

except Exception as e:
    print(f"Error in time series modeling: {e}")
    print("Using simple trend analysis instead")

    # If ARIMA fails, use simple linear regression to predict trend
    from sklearn.linear_model import LinearRegression

    # Create a numeric x-axis (days since start)
    days = np.array(range(len(daily_comp_rate))).reshape(-1, 1)
    rates = daily_comp_rate['7d_ma'].values

    # Fit linear regression
    lr_model = LinearRegression()
    lr_model.fit(days, rates)

    # Predict future trend
    future_days = np.array(range(len(daily_comp_rate), len(daily_comp_rate) + 14)).reshape(-1, 1)
    future_rates = lr_model.predict(future_days)

    # Create forecast dates
    last_date = daily_comp_rate.index[-1]
    forecast_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=14)

    # Plot linear trend forecast
    plt.figure(figsize=(14, 7))
    plt.plot(daily_comp_rate.index, daily_comp_rate['7d_ma'], label='Historical Data')
    plt.plot(forecast_dates, future_rates, label='Linear Trend Forecast', color='red')
    plt.title('Complication Rate Linear Trend Forecast')
    plt.xlabel('Date')
    plt.ylabel('Complication Rate')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
